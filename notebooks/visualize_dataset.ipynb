```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu train (chứa label và pixel ảnh)
train_path = "../dataset/train.csv"  # từ thư mục notebooks đi lên một cấp rồi vào dataset
train_df = pd.read_csv(train_path)

# Lấy 3 hàng đầu tiên
first_rows = train_df.iloc[:3]

fig, axes = plt.subplots(1, 3, figsize=(9, 3))

for ax, (_, row) in zip(axes, first_rows.iterrows()):
    label = row["label"]
    pixels = row.drop("label").to_numpy().reshape(28, 28)

    ax.imshow(pixels, cmap="gray")
    ax.set_title(f"Label: {label}")
    ax.axis("off")

# Hiển thị ảnh và in label
plt.tight_layout()
plt.show()
```
